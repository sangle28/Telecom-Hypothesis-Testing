[Calculate the monthly revenue from each user (subtract the free package limit from the total number of calls, text messages, and data; multiply the result by the calling plan value; add the monthly charge depending on the calling plan). N.B. This might not be as trivial as just a couple of lines given the plan conditions! So, it's okay to spend some time on it.]

```python
# Calculate the monthly revenue for each user (minutes surcharge)

def minutes_surcharge(df_1):
    minutes_used = df_1['minutes_per_month_per_user']
    minutes_included = df_1['minutes_included']
    minutes_difference = minutes_used - minutes_included
    
    if minutes_difference < 0:
        df_1['minutes_surcharge'] = 0
    else:
        df_1['minutes_surcharge'] = minutes_difference * df_1['usd_per_minute']
    return df_1
        
users_total_usage_and_plans = users_total_usage_and_plans.apply(minutes_surcharge, axis=1)
print(users_total_usage_and_plans[['user_id', 'month', 'minutes_per_month_per_user', 'minutes_included', 'usd_per_minute', 'minutes_surcharge']].sample(10, random_state=48))
```

```python
# Calculate the monthly revenue for each user (messages surcharge)

def messages_surcharge(df_2):
    messages_used = df_2['messages_per_month_per_user']
    messages_included = df_2['messages_included']
    messages_difference = messages_used - messages_included
    
    if messages_difference < 0:
        df_2['messages_surcharge'] = 0
    else:
        df_2['messages_surcharge'] = messages_difference * df_2['usd_per_message']
    return df_2

users_total_usage_and_plans = users_total_usage_and_plans.apply(messages_surcharge, axis=1)
print(users_total_usage_and_plans[['user_id', 'month', 'messages_per_month_per_user', 'messages_included', 'usd_per_message', 'messages_surcharge']].sample(10, random_state=48))
```

```python
# Calculate the monthly revenue for each user (gigabytes surcharge)

def data_surcharge(df_3):
    gigabytes_used = df_3['gigabytes_per_month_per_user']
    gigabytes_included = df_3['gb_per_month_included']
    gigabytes_difference = gigabytes_used - gigabytes_included
    
    if gigabytes_difference < 0:
        df_3['gigabyte_surcharge'] = 0
    else:
        df_3['gigabyte_surcharge'] = gigabytes_difference * df_3['usd_per_gb']
    return df_3

users_total_usage_and_plans = users_total_usage_and_plans.apply(data_surcharge, axis=1)
print(users_total_usage_and_plans[['user_id', 'month', 'gigabytes_per_month_per_user', 'gb_per_month_included', 'usd_per_gb', 'gigabyte_surcharge']].sample(10, random_state=48))
```

```python
# Calculate the monthly revenue for each user (total revenue)
users_total_usage_and_plans['total_revenue'] = users_total_usage_and_plans['minutes_surcharge'] + users_total_usage_and_plans['messages_surcharge'] + users_total_usage_and_plans['gigabyte_surcharge'] + users_total_usage_and_plans['usd_monthly_pay']
display(users_total_usage_and_plans.sample(10, random_state=48))
```

```python
calls['call_date'] = pd.to_datetime(calls['call_date'])

# Extracting month
calls['month'] = calls['call_date'].dt.month
calls_summary = calls.groupby(['user_id', 'month'])['duration'].sum().reset_index(name='total_duration')
users = pd.read_csv('/datasets/megaline_users.csv')[['user_id', 'plan']]
calls_summary_with_plan = calls_summary.merge(users, on='user_id', how='left')
average_call_duration = calls_summary_with_plan.groupby(['plan','month'])['total_duration'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=average_call_duration, x='month', y='total_duration', hue='plan')
plt.title('Average Duration of Calls per Plan per Month')
plt.xlabel('Month')
plt.ylabel('Average Call Duration (minutes)')
plt.legend(title='Plan')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.histplot(data=calls_summary_with_plan, x='total_duration', hue='plan', bins=30, kde=True)

handles, labels = [], []
unique_plans = calls_summary_with_plan['plan'].unique()

# Generate a color palette for the number of unique plans
palette = sns.color_palette(n_colors=len(unique_plans))

for idx, plan in enumerate(unique_plans):
    handles.append(plt.Rectangle((0, 0), 1, 1, color=palette[idx]))
    labels.append(plan)

plt.legend(handles=handles, labels=labels, title='Plan')
plt.title('Total Call Duration per Plan per Month')
plt.xlabel('Total Call Duration (minutes)')
plt.ylabel('Frequency')

plt.show()
```
